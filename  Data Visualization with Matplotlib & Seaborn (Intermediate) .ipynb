```python
import matplotlib.pyplot as plt

# Dummy data
x = [1, 2, 3, 4, 5]
y = [10, 15, 8, 12, 18]

# ----------------------
# Line Chart
# ----------------------
plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Simple Line Chart')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True)
plt.show()

# ----------------------
# Bar Chart
# ----------------------
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

plt.figure(figsize=(8, 4))
plt.bar(categories, values, color='orange')
plt.title('Simple Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

# 1. Histogram – Visualize Numeric Data Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dummy numeric data
data = {
    'ages': [22, 25, 30, 35, 40, 22, 30, 28, 35, 45, 50, 22, 25]
}
df = pd.DataFrame(data)

# Plot histogram
plt.figure(figsize=(8, 4))
plt.hist(df['ages'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# 2. Pie Chart – Visualize Categorical Data Distribution

```python
# Dummy categorical data
category_data = {
    'Category': ['Apples', 'Bananas', 'Cherries', 'Dates'],
    'Count': [30, 45, 15, 10]
}
cat_df = pd.DataFrame(category_data)

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(cat_df['Count'], labels=cat_df['Category'], autopct='%1.1f%%', startangle=90, colors=['red', 'yellow', 'pink', 'brown'])
plt.title('Fruit Distribution')
plt.axis('equal')  # Ensures pie is a circle
plt.show()
```

# Example 2: Using Real Data (e.g., Titanic Dataset)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
df = pd.read_csv('Titanic-Dataset.csv')

# Remove rows with missing values in selected columns
df_clean = df[['Age', 'Fare']].dropna()

# Scatter plot: Age vs Fare
plt.figure(figsize=(8, 5))
plt.scatter(df_clean['Age'], df_clean['Fare'], alpha=0.6, color='blue')
plt.title('Age vs Fare (Titanic Dataset)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

# 1. Heatmap – Correlation Matrix

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = pd.read_csv('Titanic-Dataset.csv')

# Select numerical columns
numeric_df = df.select_dtypes(include='number')

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap (Titanic Dataset)')
plt.show()
```

# 2. Pair Plot – Pairwise Relationships

```python
# Choose a small subset of numerical columns
subset = df[['Age', 'Fare', 'Pclass', 'Survived']].dropna()

# Plot pair plot
sns.pairplot(subset, hue='Survived', palette='husl')  # `hue` adds color based on Survived
plt.suptitle('Pair Plot (Titanic Subset)', y=1.02)
plt.show()
```

# 1. Histogram – Age Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# 2. Bar Chart – Count of Passengers by Class

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', data=df, palette='pastel')
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

# 3. Pie Chart – Gender Distribution

```python
gender_counts = df['Sex'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

# 4. Scatter Plot – Age vs Fare

```python
df_clean = df[['Age', 'Fare']].dropna()

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Fare', data=df_clean)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

# 5. Box Plot – Fare by Class

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

# 6. Heatmap – Correlation Between Numeric Features

```python
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

# 7. Pair Plot – Pairwise Comparison

```python
subset = df[['Age', 'Fare', 'Pclass', 'Survived']].dropna()
sns.pairplot(subset, hue='Survived', palette='Set1')
plt.suptitle('Pair Plot of Titanic Subset', y=1.02)
plt.show()
```
